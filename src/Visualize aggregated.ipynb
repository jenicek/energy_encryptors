# Visualize aggregated

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

dataset = pd.read_csv('smart_meters_london_2013.csv')
dataset = dataset.T
```

```python
mean = dataset.iloc[1:,:].mean(axis=1)
std = dataset.iloc[1:,:].std(axis=1)
plt.scatter(mean, std)
plt.xlabel("mean")
plt.ylabel("std")
plt.title("sensor statistics")
```
